**Étape 3 : Faire une requête à l'API**

Une fois que vous avez votre clé API et les dépendances installées, vous pouvez commencer à utiliser l'API COVID-19 Stats. Voici comment vous pouvez faire une requête pour obtenir des statistiques COVID-19 par pays :

```python
import requests

# Remplacez "votre_clé_api" par votre clé API obtenue sur RapidAPI

url = "https://covid-193.p.rapidapi.com/statistics"

headers = {
	"X-RapidAPI-Key": "dbeaa5b72cmsha913704fa16b6fep1f851djsn63de81f4ee75",
	"X-RapidAPI-Host": "covid-193.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

# Vérifiez si la requête a réussi
if response.status_code == 200:
    # Afficher les données JSON de réponse
    data = response.json()
    print(data)
else:
    print("Erreur lors de la requête :", response.status_code)
```

**Étape 4 : Analyser les données**

Une fois que vous avez récupéré les données, vous pouvez les analyser selon vos besoins. Par exemple, vous pouvez extraire les statistiques pour chaque pays et les afficher

```python
if response.status_code == 200:
    data = response.json()
    for country_data in data["response"]:
        country_name = country_data["country"]
        total_cases = country_data["cases"]["total"]
        total_deaths = country_data["deaths"]["total"]
        total_recovered = country_data["cases"]["recovered"]
        print(f"Pays : {country_name}, Cas total : {total_cases}, Décès total : {total_deaths}, Rétablis : {total_recovered}")
else:
    print("Erreur lors de la requête :", response.status_code)
```
